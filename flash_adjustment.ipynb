```python
from flashAdj import *
import matplotlib.pyplot as plt
```

```python
amb_img = cv2.imread('./m_dataset/bottle2_ambient (1).jpg')
flash_img = cv2.imread('./m_dataset/bottle2_flash (1).jpg')
a = [-0.5, 0, 0.25, 0.5, 1, 1.5]
fig = plt.figure(figsize=([12, 12]))
j = 1
for i in a:
    out = flashAdj(amb_img, flash_img, i)
    out[out>255]= 255
    fig.add_subplot(3,2,j)
    j = j+1
    plt.title('alpha = ' + str(i))
    plt.imshow(out)
```
